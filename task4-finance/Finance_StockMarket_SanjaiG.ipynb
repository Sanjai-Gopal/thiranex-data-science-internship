## Objective
End-to-end analysis of stock market data for 5 companies across 2 years (2023–2024).

### Steps:
1. Dataset overview (2,610 daily records × 5 stocks)
2. Price trend analysis with moving averages
3. Risk vs Return analysis & Sharpe Ratio
4. Volatility & correlation analysis
5. Monthly seasonality heatmap
6. 9-chart dashboard with investment insights


```
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
plt.style.use('dark_background')
print("✅ Libraries ready")
```

```
# Load Dataset
df = pd.read_csv('stock_market_data.csv', parse_dates=['Date'])
print("Shape:", df.shape)
print("\nDate Range:", df['Date'].min(), "→", df['Date'].max())
print("\nCompanies:", df['Ticker'].unique())
print("\nSample:")
df.head()
```

```
# Summary Statistics per Stock
summary = df.groupby('Ticker').agg(
    Avg_Close=('Close','mean'),
    Min_Close=('Close','min'),
    Max_Close=('Close','max'),
    Avg_Volume=('Volume','mean'),
    Avg_Daily_Return=('Daily_Return','mean'),
    Std_Daily_Return=('Daily_Return','std'),
).round(4)
summary['Annual_Return_%'] = (summary['Avg_Daily_Return']*252*100).round(2)
summary['Annual_Volatility_%'] = (summary['Std_Daily_Return']*np.sqrt(252)*100).round(2)
summary['Sharpe_Ratio'] = (summary['Annual_Return_%']/summary['Annual_Volatility_%']).round(3)
display(summary.drop(['Avg_Daily_Return','Std_Daily_Return'], axis=1))
```

```
# 2-Year Total Returns
print("=== 2-YEAR TOTAL RETURNS ===")
for ticker in df['Ticker'].unique():
    sub = df[df['Ticker']==ticker].sort_values('Date')
    total_return = (sub['Close'].iloc[-1]/sub['Close'].iloc[0] - 1)*100
    print(f"  {ticker:12s}: {total_return:+.1f}%")
    
print("\n=== BEST TRADING DAYS ===")
best = df.nlargest(5,'Daily_Return')[['Date','Ticker','Close','Daily_Return']]
print(best.to_string(index=False))

print("\n=== WORST TRADING DAYS ===")
worst = df.nsmallest(5,'Daily_Return')[['Date','Ticker','Close','Daily_Return']]
print(worst.to_string(index=False))
```
